```python
import pandas as pd

df = pd.read_csv(r"C:\Users\User\Downloads\Triangel\CarSales.csv")
df = df.drop_duplicates()
df = df.drop(columns=['Vehicle Type'])
df['Model'] = df['Model'].str.replace('05-Sep', '9-5').str.replace('03-Sep', '9-3')
df = df.dropna(subset=['Year Resale Value']).dropna(subset=['Price in Thousands'])
df = df.dropna(subset=['Curb Weight']).dropna(subset=['Power Factor']).dropna(subset=['Fuel Efficiency'])
def format_currency(value):
    if pd.isna(value):
        return "N/A"  # Handle missing values
    amount = value * 1000  # Convert from thousands to actual value
    return f"${amount:,.0f}"  # Format as currency with commas

# Apply formatting to the three columns
df['Sales in Thousands'] = df['Sales in Thousands'].apply(format_currency)
df['Year Resale Value'] = df['Year Resale Value'].apply(format_currency)
df['Price in Thousands'] = df['Price in Thousands'].apply(format_currency)
df['Horsepower'] = df['Horsepower'].astype(int)
df['Fuel Efficiency'] = df['Fuel Efficiency'].astype(int)
df['Curb Weight'] = df['Curb Weight'].round(1)
df['Power Factor'] = df['Power Factor'].round(1)
df['Sales'] = df['Sales in Thousands']
df
```
